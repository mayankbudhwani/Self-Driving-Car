```python
def pipeline(img):
    rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
    height,width,channel=img.shape
    #FOR YELLOW
    lt=np.uint8([15,38,115])
    ht=np.uint8([35,204,255])
    ymask=cv2.inRange(img,lt,ht)
    
    #FOR WHITE
    lt=np.uint8([0,200,0])
    ht=np.uint8([180,255,255])
    wmask=cv2.inRange(img,lt,ht)
    
    mask=cv2.bitwise_or(ymask,wmask)
    ywimage=cv2.bitwise_and(rgb,rgb,mask=mask)
    
    gray=cv2.cvtColor(ywimage,cv2.COLOR_RGB2GRAY)
    
    kernel_size=5
    gbgray=cv2.GaussianBlur(gray, (kernel_size, kernel_size), 0) 
    
    canny=cv2.Canny(gbgray,50,150)
    
    im=np.zeros_like(canny,np.uint8)
    boundary=np.array([[[int(width/10)*2,(int(height/10)*10)-1],[int(width/10)*5,int(height/10)*5],[int(width/10)*7,int(height/10)*5],[int(width/10)*9-30,(int(height/10)*10)-1]]],'int32')#REctangular Boundaries
    regionmask=cv2.fillPoly(im,boundary,(255))#Black and White mask for AND operation
    region=cv2.bitwise_and(im,canny,mask=regionmask)
    
    lines=cv2.HoughLinesP(region,1,np.pi/180,15,20,10)
    
    
    middle_x=int(width/2)
    right_lines=[]
    left_lines=[]
    for l in lines:
        
        x1=l[0][0]
        x2=l[0][2]

        y1=l[0][1]
        y2=l[0][3]
        
        if(y2-y1==0 or x2-x1==0):
            continue
        m=y2-y1/x2-x1
        
        epsilon = 0.1
        if abs(m) <= epsilon:
            continue
        
        if m < 0 and x1 < middle_x and x2 < middle_x:
            left_lines.append(l)
            #cv2.line(rgb, (x1,y1), (x2,y2), (255,0,0), 5, cv2.LINE_AA)
        elif x1 >= middle_x and x2 >= middle_x:
            right_lines.append(l)
            #cv2.line(rgb, (x1,y1), (x2,y2), (0,0,255), 5, cv2.LINE_AA)
    #print("New")
    #print(right_lines)
    #print()
    #print(left_lines,sep="\n")
    if(left_lines==[] or right_lines==[]):
        return rgb,1
    from scipy import stats
    def equation(lines):
        xs=[]
        ys=[]
        for l in lines:
            xs.append(l[0][0])
            xs.append(l[0][2])

            ys.append(l[0][1])
            ys.append(l[0][3])
        m,c,_,_,e=stats.linregress(xs,ys)

        return m,c
    def getx(y,c,m):
        return int((y-c)/m)
    
    mr,cr=equation(right_lines)
    ml,cl=equation(left_lines) 
    
    if(mr<0 or ml>0):
        return rgb,1
    cv2.line(rgb, (getx(int(height/10)*10-1,cr,mr),int(height/10)*10-1), (getx(int(height/10)*6,cr,mr),int(height/10)*6), (255,0,0), 8, cv2.LINE_AA)
    
    cv2.line(rgb, (getx(int(height/10)*10-1,cl,ml),int(height/10)*10-1), (getx(int(height/10)*6,cl,ml),int(height/10)*6), (0,0,255), 8, cv2.LINE_AA)
    
    
    
    return rgb,0
```

```python

# For  VIDEO
#Memory Function is still incomplete, slopes and intercepts of current line must follow previous lines (can be solved by mean or gradient)
#We can detect curves instead of lines for curved roads
#This is not performing well in shadow (Unprocessed Frames)
cap = cv2.VideoCapture('challenge.mp4')

fourcc = cv2.VideoWriter_fourcc(*'DIVX')
out = cv2.VideoWriter('output1.mp4',fourcc, 20.0, (1280,720),True)
Unprocessed=0
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret==True:
        count+=1
        ans,up=pipeline(frame)
        if(up==1):
            Unprocessed+=1
        out.write(ans)
        #cv2.imshow('frame',ans)
        #if(count%20==0):
            #cv2.imwrite("ch%s.jpg"%count,ans)
        if cv2.waitKey(20) & 0xFF == ord('q'):
            break
    else:
        break
print(Unprocessed)
out.release()
cap.release()
cv2.destroyAllWindows()
```
